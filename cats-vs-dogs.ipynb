```python
#loading dataset

from datasets import load_dataset

cvd_dataset=load_dataset("cats_vs_dogs")
print(cvd_dataset["train"][0])
```

```python
#transforming pil to numpy

import numpy as np

cvd_dict={
    'images':[],
    'labels':[]
}

for sample in cvd_dataset['train']:
    img=np.array(sample['image'])
    lbl=np.array(sample['labels'])

    cvd_dict['images'].append(img)
    cvd_dict['labels'].append(lbl)
```
